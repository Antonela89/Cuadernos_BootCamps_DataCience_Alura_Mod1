```python
import pandas as pd
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'),list('ZYX'))
df
```

```python
df.sort_index(inplace = True) #ordena ascentede los indices
df
```

```python
df.sort_index(inplace = True, axis = 1) #axis = 1 ordena las columnas
df
```

```python
df.sort_values(by ='X', inplace = True) #pide una columna para ordenar de forma ascendente
df
```

```python
df.sort_values(by = '3',axis = 1, inplace = True) #pide una fila para ordenar de forma ascendente
df
```

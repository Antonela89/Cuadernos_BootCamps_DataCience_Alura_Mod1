```python
import pandas as pd
```

#Series

```python
data = [1,2,3,4,5]
```

```python
s = pd.Series(data)
s
```

```python
index = ['Linea' + str(i) for i in range(5)]
index
```

```python
s = pd.Series(data = data, index = index)
s
```

```python
data = {'Linea' + str(i): i+1 for i in range(5)}
data
```

```python
s = pd.Series(data)
s
```

```python
s1 = s + 2
s1
```

```python
s2 = s1 + s #cuando ambas series tiene el mismo indice puede operar entre ellas
s2
```

Ejemplos

```python
datos = {'A': {'X': 1, 'Y': 3}, 'B': {'X': 2, 'Y': 4}}
df = pd.DataFrame(datos)
df
```

```python
datos = [('A', 'B'), ('C', 'D')]
df = pd.DataFrame(datos, columns = ['L1', 'L2'],  index = ['C1', 'C2'])
df
```

#Dataframe

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
df1 = pd.DataFrame(data = data)
df1
```

```python
index = ['Línea' + str(i) for i in range(3)]
index
```

```python
df1 = pd.DataFrame(data = data, index = index)
df1
```

```python
columns = ['Columna' + str(i) for i in range(3)]
columns
```

```python
df1 = pd.DataFrame(data = data, index = index, columns = columns)
df1
```

```python
data =  {'Columna0': {'Línea0': 1, 'Línea1': 4, 'Línea2': 7},
         'Columna1': {'Línea0': 2, 'Línea1': 5, 'Línea2': 8},
         'Columna2': {'Línea0': 3, 'Línea1': 6, 'Línea2': 9}}
data
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
data2 = [(1,2,3),(4,5,6),(7,8,9)]
data2
```

```python
df3 = pd.DataFrame(data2, index = index, columns = columns)
df3
```

```python
df1[df1 > 0] = 'A'
df1
```

```python
df2[df2 > 0] = 'B'
df2
```

```python
df3[df3 > 0] = 'C'
df3
```

```python
df4 = pd.concat([df1,df2,df3]) #uso la columna como llave (concatencacion vertical) y las lineas se usaron como secuencia
df4
```

```python
df5 = pd.concat([df1, df2, df3], axis = 1) #uso la linea como llave (concatencacion horizontal) y las columnas se usaron como secuencia
df5
```
